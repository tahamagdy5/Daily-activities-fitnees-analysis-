```python
plt.scatter(df['TotalSteps'],df['Calories'],alpha=0.5)
median_steps=5986
median_calories=2062
plt.axhline(median_calories,color = 'red' , label ='median of the calroies')
plt.axvline(median_steps , color ='green', label ='median of the steps')
plt.legend()
plt.xlabel('steps')
plt.ylabel('calories')
plt.title('calories by total steps')
plt.show()
```
